```python
p = pageObj.extract_text()
```

```python
liste = p.splitlines()
```

```python
liste
```

```python
liste_propre = liste[4:37]
```

```python
liste_propre
```

```python
# Supprimer les espaces supplémentaires
liste2 = [element.strip() for element in liste_propre]
```

```python
liste2
```

```python
import pandas as pd
import re
```

```python
# Créer une liste avec les colonnes du dataframe
colonnes = ['Gas', 'GWP (100-year)']

# Créer une liste de listes avec les données
donnees = []

for element in liste_propre:
    # Utiliser une expression régulière pour diviser la ligne en substance et GWP
    match = re.match(r'(\S+)\s+(\S+)', element)
    if match:
        substance = match.group(1)
        gwp = match.group(2)
        donnees.append([substance, gwp])

# Créer le dataframe
df = pd.DataFrame(donnees, columns=colonnes)
```

```python
df
```

```python
df = df.drop(0)
```

```python
df = df.reset_index()
```

```python
del df['index']
```

```python
df.at[0, 'Gas'] = "CO2"
df.at[0, 'GWP (100-year)'] = "1"

df.at[1, 'Gas'] = "CH4"
df.at[1, 'GWP (100-year)'] = "25"
```
